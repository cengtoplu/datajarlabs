```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mu = 10
sigma = 5
n = 10000

dagilim = np.random.normal(mu, sigma, n)
```

```python
dagilim
```

```python
plt.figure(figsize=(10,4), dpi = 200)
plt.hist(dagilim, bins = 100)
plt.show()
```

```python
dagilim.sort()
```

```python
dagilim
```

```python
from scipy import stats
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4), dpi = 200)
nsample = 100
res = stats.probplot(dagilim, plot=plt)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.title("Q-Q Çizgesi", fontdict = baslik_font)
plt.xlabel("Teorik Değerler", fontdict = eksen_font)
plt.ylabel("Örnek Veriler", fontdict = eksen_font)
plt.show()
```

Not: https://finance.yahoo.com/ adresinden veri şu şekilde çekilir. Verilen adres tıkladıktan sonra arama bölümüne aramak istediğiniz hisse senedinin adını yazın (ör: google). Arama sonucu altta çıkan sonuca tıklayarak ilgili hisse senedine gidiniz. İlgili hisse senedinin bilgilerinin bulunduğu sayfaya geldiğinizde Historical Prices (Geçmiş veriler) sekmesine tıklayın ve ilgili tarih aralığını seçtikten sonra Apply'a tıklayın.
```python
goog = pd.read_csv('google.csv')
```

```python
goog
```

```python
goog['Getiriler'] = goog['Close'].diff()
print(goog.head())
```

```python
mu, sigma, n = 0.5, 0.1, 504
normaldagilimli = np.random.normal(mu, sigma, n)
```

```python
getiriler = list(goog['Getiriler'])
```

```python
print(getiriler)
```

```python
print(len(getiriler))
```

```python
goog.describe()
```

```python
normaldagilim = np.random.normal (np.mean(goog['Getiriler']), np.std(goog['Getiriler']), 505)
```

```python
getiriler.sort()
normaldagilim.sort()

plt.figure(figsize=(8,5), dpi = 200)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
plt.title('Getiriler & Normal Dağılımlı Veri', fontdict = baslik_font)
plt.xlabel('Getiriler', fontdict = eksen_font)
plt.ylabel('Normal Dağılımlı Veri', fontdict = eksen_font)
plt.scatter(normaldagilim, getiriler)
plt.plot([np.min(normaldagilim), np.max(normaldagilim)],
         [np.min(getiriler), np.max(getiriler)], c='red')
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4), dpi = 200)
nsample = 100
res = stats.probplot(getiriler, plot=plt)
plt.title("Q-Q Çizgesi", fontdict = baslik_font)
plt.xlabel("Teorik Değerler", fontdict = eksen_font)
plt.ylabel("Örnek Veriler", fontdict = eksen_font)
plt.show()
```
